```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Data
x = np.array([5, 15, 25, 35, 45, 55]).reshape((-1, 1))
y = np.array([5, 20, 14, 32, 22, 38])

# Linear Regression Model
model = LinearRegression()
model.fit(x, y)

# Predicted values
y_pred = model.predict(x)

# Plotting
plt.scatter(x, y, color='green', label='Actual response, $y_i$')  # Actual data points
plt.scatter(x, y_pred, color='red', marker='s', label='Predicted response, $f(x_i) = b_0 + b_1x_i$')  # Predicted points
plt.plot(x, y_pred, color='black', label='Estimated regression line, $f(x) = b_0 + b_1x$')  # Regression line

# Residuals
for i in range(len(x)):
    plt.plot([x[i], x[i]], [y[i], y_pred[i]], 'k--', color='gray')  # Residual lines

# Labels and legend
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline

# Given dataset
x = np.array([5, 15, 25, 35, 45, 55]).reshape((-1, 1))
y = np.array([5, 20, 14, 32, 22, 38])

# Degrees of polynomial regression to test
degrees = [1, 2, 3, 5]

# Create a figure for subplots
fig, axes = plt.subplots(2, 2, figsize=(10, 10))
axes = axes.ravel()

# Generate plots for each polynomial degree
for i, degree in enumerate(degrees):
    ax = axes[i]
    
    # Create polynomial regression model
    poly = PolynomialFeatures(degree)
    x_poly = poly.fit_transform(x)
    
    model = LinearRegression()
    model.fit(x_poly, y)
    y_pred = model.predict(x_poly)
    
    # Calculate R² score correctly
    r2_score = model.score(x_poly, y)
    
    # Scatter plot of actual data
    ax.scatter(x, y, color='green', label='Actual response, $y_i$')
    
    # Plot the polynomial regression curve
    x_curve = np.linspace(x.min(), x.max(), 100).reshape(-1, 1)
    x_curve_poly = poly.transform(x_curve)
    y_curve = model.predict(x_curve_poly)
    
    ax.plot(x_curve, y_curve, color='black', label='Estimated regression line, $f(x)$')
    
    # Labels and legend
    ax.set_xlabel('$x$')
    ax.set_ylabel('$y$')
    ax.legend()
    ax.set_title(f'Degree: {degree}, $R^2$ = {r2_score:.2f}')
    ax.grid(True)

# Adjust layout and display plot
plt.tight_layout()
plt.show()
```

Polynomial Regression With scikit-learn

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
x = np.array([5, 15, 25, 35, 45, 55]).reshape((-1, 1))
y = np.array([15, 11, 2, 8, 25, 32])
```

```python
transformer = PolynomialFeatures(degree=2, include_bias=False)
```

```python
transformer.fit(x)
```

```python
x_ = transformer.transform(x)
```

```python
x_ = PolynomialFeatures(degree=2, include_bias=False).fit_transform(x)
```

```python
x_
```

```python
model = LinearRegression().fit(x_, y)
```

```python
r_sq = model.score(x_, y)
print(f"coefficient of determination: {r_sq}")
```

```python
print(f"intercept: {model.intercept_}")
```

```python
print(f"coefficients: {model.coef_}")
```

```python

x_ = PolynomialFeatures(degree=2, include_bias=True).fit_transform(x)
```

```python
x_
```

```python
model = LinearRegression(fit_intercept=False).fit(x_, y)
```

```python
r_sq = model.score(x_, y)
print(f"coefficient of determination: {r_sq}")
```

```python
print(f"intercept: {model.intercept_}")
```

```python
print(f"coefficients: {model.coef_}")
```

```python
y_pred = model.predict(x_)
print(f"predicted response:\n{y_pred}")
```

Advanced Linear Regression With statsmodels

```python
import numpy as np
import statsmodels.api as sm
```

```python
x = [
  [0, 1], [5, 1], [15, 2], [25, 5], [35, 11], [45, 15], [55, 34], [60, 35]
]
y = [4, 5, 20, 14, 32, 22, 38, 43]
x, y = np.array(x), np.array(y)
```

```python
x = sm.add_constant(x)
```

```python
x
```

```python
y
```

```python
model = sm.OLS(y, x)
```

```python
results = model.fit()
```

```python
print(results.summary())
```

```python
print(f"coefficient of determination: {results.rsquared}")
```

```python
print(f"adjusted coefficient of determination: {results.rsquared_adj}")
```

```python
print(f"regression coefficients: {results.params}")
```

```python
print(f"predicted response:\n{results.fittedvalues}")
```

```python
print(f"predicted response:\n{results.predict(x)}")
```

```python
x_new = sm.add_constant(np.arange(10).reshape((-1, 2)))
x_new
```

```python
y_new = results.predict(x_new)
y_new
```
